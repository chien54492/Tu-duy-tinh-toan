```python
ten = input("Ten chu ho: ")
truoc = int(input("Chi so thang truoc: "))
nay = int(input("Chi so thang nay: "))

kwh = nay - truoc
tien = 0

if kwh <= 50:
    tien = kwh * 1984
elif kwh <= 100:
    tien = 50 * 1984 + (kwh - 50) * 2050
elif kwh <= 200:
    tien = 50 * 1984 + 50 * 2050 + (kwh - 100) * 2380
elif kwh <= 300:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (kwh - 200) * 2998
elif kwh <= 400:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (kwh - 300) * 3350
else:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (kwh - 400) * 3460

tien = tien * 1.08

print("Ho va ten:", ten)
print("Tien phai tra la:", int(tien))
```
