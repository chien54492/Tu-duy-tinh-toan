bài 2
Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu
vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần
diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2
chữ số phần thập phân.

```python
a = float(input("Nhap chieu rong a: "))
b = float(input("Nhap chieu dai b: "))

pi = 3.14
r = min(a, b) / 2

dien_tich_hcn = a * b
dien_tich_tron = pi * r * r

ket_qua = dien_tich_hcn - dien_tich_tron

print("Dien tich con lai:", round(ket_qua, 2))
```

bài 6
Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.

```python
import math

a = float(input("Nhap a: "))
b = float(input("Nhap b: "))
c = float(input("Nhap c: "))

if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    s = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print("Dien tich:", round(s, 1))
else:
    print("Khong phai 3 canh cua tam giac")
```

Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này,
tính tiền điện tháng này cho hộ biết:
- Từ 0 – 50kWh: 1.984 đồng/kWh
- Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh

```python
ten = input("Ten chu ho: ")
truoc = int(input("Chi so thang truoc: "))
nay = int(input("Chi so thang nay: "))

kwh = nay - truoc
tien = 0

if kwh <= 50:
    tien = kwh * 1984
elif kwh <= 100:
    tien = 50 * 1984 + (kwh - 50) * 2050
elif kwh <= 200:
    tien = 50 * 1984 + 50 * 2050 + (kwh - 100) * 2380
elif kwh <= 300:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (kwh - 200) * 2998
elif kwh <= 400:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (kwh - 300) * 3350
else:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (kwh - 400) * 3460

tien = tien * 1.08

print("Ho va ten:", ten)
print("Tien phai tra la:", int(tien))
```

bài 9
Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số
tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case

```python
num = int(input("Nhập số từ 0 đến 9: "))

match num:
    case 0:
        print("Không")
    case 1:
        print("Một")
    case 2:
        print("Hai")
    case 3:
        print("Ba")
    case 4:
        print("Bốn")
    case 5:
        print("Năm")
    case 6:
        print("Sáu")
    case 7:
        print("Bảy")
    case 8:
        print("Tám")
    case 9:
        print("Chín")
    case _:
        print("Số không hợp lệ")
```
