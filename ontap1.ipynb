14Viết hàm trả về hệ số nhị thức C(n, k). 

```python
def C(n, k):
    if k < 0 or k > n:
        return 0
    
    k = min(k, n - k)  # tối ưu
    res = 1
    
    for i in range(1, k + 1):
        res = res * (n - i + 1) // i
    
    return res

# Nhập
n, k = map(int, input("Nhập n k: ").split())
print(C(n, k))
```

16 Viết hàm trả về góc ∠ABC theo đơn vị độ khi biết tọa độ các điểm A(x1,y1), B(x2,y2), C(x3,y3). 

```python
import math
def goc_ABC(x1, y1, x2, y2, x3, y3):
    A,B=x2-x1,y2-y1
    A,C=x3-x1,y3-y1
    d=A*C-A*B
    lenAB=math.sqrt(A**2 + B**2)
    lenAC=math.sqrt(A**2 + C**2)
    if lenAB==0 or lenAC==0:
        return None
    cos_theta=d/(lenAB*lenAC)
    cos_theta = max(-1, min(1, cos_theta))
    goc=math.acos(cos_theta)
    return math.degrees(goc)
x1,y1=(map(int, input().split()))
x2,y2=(map(int, input().split()))
x3,y3=(map(int, input().split()))
print(goc_ABC(x1, y1, x2, y2, x3, y3))
  

    

```

chấm động, ngược lại trả về 'neither'.
- Đầu vào: chuỗi s.
- Đầu ra: một trong các giá trị {'int', 'float', 'neither'}.
- Quy tắc:
- Cho phép có dấu + hoặc - ở đầu.
- Số nguyên phải chỉ gồm các chữ số và không có dấu chấm thập phân (ví dụ: '100').
- Số thực phải có dấu chấm và có chữ số ở cả hai phía của dấu chấm (ví dụ: '100.0', '-3.14').
- Các chuỗi như '.5', '5.', '1e3' được xem là 'neither' trong bài này.


```python
def detect_number_type(s):
    if s[0]=="+" or s[0]=="-":
        s=s[1:]
    if not s:
        return "neither"
    if s.isdigit():
        return "int"
    elif s.count(".")==1:
        left, right = s.split(".")
        if left.isdigit() and right.isdigit():
          return "float"
    else:
        return "neither"
s=input()
print(detect_number_type(s))
```

- Đầu ra: giá trị boolean.
- Lưu ý: Số âm không phải là số chính phương trong tập số nguyên. 

```python
def is_perfect_square(n):
    if n<0:
        return False
    for i in range(1, int(n**0.5)+1):
        if i*i==n:
            return True
    return False
```

Giữ nguyên chữ hoa/chữ thường và quay vòng trong phạm vi A..Z hoặc a..z. 

```python
def shift_char(ch, n):
    if not isinstance(ch, str) or len(ch) != 1:
        raise ValueError("Đầu vào phải là đúng 1 ký tự")

    if 'A' <= ch <= 'Z':
        start = ord('A')
        return chr((ord(ch) - start + n) % 26 + start)

    elif 'a' <= ch <= 'z':
        start = ord('a')
        return chr((ord(ch) - start + n) % 26 + start)

    else:
        raise ValueError("Ký tự phải là chữ cái a-z hoặc A-Z")
```

43 Viết hàm tính diện tích tam giác có ba cạnh a, b, c (dùng công thức Heron). Phát sinh ValueError nếu ba cạnh không thể
tạo thành tam giác

```python
def dien_tich(a,b,c):
    s=(a+b+c)/2
    if a+b>c and a+c>b and b+c>a:
        return (s*(s-a)*(s-b)*(s-c))**0.5
    else:
        return "ValueError"  
```

44 Viết hàm trả về tích các số nguyên từ 1 đến n (giai thừa). Với n=0, trả về 1. 

```python
def tich(n):
    s=1
    for i in range(1, n+1):
        s*=i
    return s
n=int(input())
print(tich(n))
```

45 Viết hàm trả về số lượng số nguyên dương k sao cho 1 <= k < n và k > (n - 1) / 2. (Ví dụ: n=9 -> 4

```python
def so_luong(n):
    s=0
    for i in range(1, n+1):
        if 1<=i<n and i>(n-1)/2:
            s+=1
    return s
n=int(input())
print(so_luong(n))
```
