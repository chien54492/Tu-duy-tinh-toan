W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
def numbertype(s):
    if s[0]in '+-':
        s=s[1:]
    if not s:
        return "neither"
    if s.isdigit():
        return "int"
    t=0
    for i in s:
        if i=='.':
            t+=1
        elif not i.isdigit():
            return "neither"
    if t>1:
        return "neither"
    if t==0:
        return "int"
    p=s.split('.')
    if p[0]==''and p[1]=='':
        return "neither"
    return "float"

s=input()
            
print(numbertype(s))
```

Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),
(0,3)], p = (2,1)

```python
def area(a, b, c):
    return abs((a[0]*(b[1]-c[1]) +
                b[0]*(c[1]-a[1]) +
                c[0]*(a[1]-b[1])) / 2)

def point_in_triangle(a, b, c, p):
    A = area(a, b, c)
    A1 = area(p, b, c)
    A2 = area(a, p, c)
    A3 = area(a, b, p)
    
    return abs(A - (A1 + A2 + A3)) < 1e-9

def point_in_quadrilateral(quad, p):
    A, B, C, D = quad
    
    return (point_in_triangle(A, B, C, p) or
            point_in_triangle(A, C, D, p))
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math
def goc_ABC(x1, y1, x2, y2, x3, y3):
    A,B=x2-x1,y2-y1
    A,C=x3-x1,y3-y1
    d=A*C-A*B
    lenAB=math.sqrt(A**2 + B**2)
    lenAC=math.sqrt(A**2 + C**2)
    if lenAB==0 or lenAC==0:
        return None
    cos_theta=d/(lenAB*lenAC)
    cos_theta = max(-1, min(1, cos_theta))
    goc=math.acos(cos_theta)
    return math.degrees(goc)
x1,y1=(map(int, input().split()))
x2,y2=(map(int, input().split()))
x3,y3=(map(int, input().split()))
print(goc_ABC(x1, y1, x2, y2, x3, y3))
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0.

```python
def sum_1_to_n(n):
    if n <= 0:
        return 0
    
    total = 0
    
    for x in range(1, n + 1):
        s = str(x)
        if len(s) == len(set(s)):
            total += x
    
    return total
```
