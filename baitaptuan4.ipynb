[PrimeNumber]
VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi.
Kiểm tra số này có phải là số nguyên tố hay không?

```python
while True:
    n = int(input("Nhập một số nguyên dương: "))
    if n > 0:
        break  
    print("Vui lòng nhập lại (n phải > 0)!")
is_prime = True 
if n <= 1:
    is_prime = False
elif n == 2:
    is_prime = True
elif n % 2 == 0:
    is_prime = False
else: 
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            is_prime = False  
            break     
        else:
            is_prime = True       
if is_prime:
    print(f"{n} là số nguyên tố")
else:
    print(f"{n} không phải là số nguyên tố")
```

W4A7
"[LargestPrimeFactor]
Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n."

```python
n = int(input())

max_prime = 0

i = 2
while i * i <= n:
    while n % i == 0:
        max_prime = i
        n //= i
    i += 1


if n > 1:
    max_prime = n

print(max_prime)
```

W4A8
ReverseAddPalindrome]
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.

```python
n = int(input())

steps = 0

while True:
    s = str(n)
    if s == s[::-1]:
        break
    r= int(s[::-1])
    n = n + r
    steps += 1

print(steps)
print(n)
```

W4A9
[PerfectSquaresDistinctDigits]
Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu
cách). Nếu không có, in ra (no number)

```python
n = int(input())

found = False

for i in range(1, int(n**0.5) + 1):
    x = i * i
    s = str(x)
    
    ok = True
    for j in range(len(s)):
        for k in range(j + 1, len(s)):
            if s[j] == s[k]:
                ok = False
                break
        if not ok:
            break
    
    if ok:
        print(x, end=" ")
        found = True

if not found:
    print("(no number)")
```

W4A10
"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn sốnhỏ nhất."

```python
n = int(input())

max_len = 0
result_x = 1

for x in range(1, n + 1):
    temp = x
    length = 1   # tính cả số ban đầu
    
    while temp != 1:
        if temp % 2 == 0:
            temp = temp // 2
        else:
            temp = temp * 3 + 1
        length += 1
    
    if length > max_len:
        max_len = length
        result_x = x

print(result_x, max_len)
```

[ClosePairNumber]
Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật:
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
Ví dụ 220 và 284 là một cặp số thân thiết
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55,
110. Tổng của 11 ước này cho chúng ta 284
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng
vừa đúng bằng 220
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
a=int(input("Nhập một số nguyên dương: "))
b=int(input("Nhập một số nguyên dương: "))
c=0
d=0
for i in range(1, a):
    if a%i==0:
        c+=i
for j in range(1, b):
    if b%j==0:
        d+=j
if c==b and a==d:
    print("True")
else:    
    print("False")
```
