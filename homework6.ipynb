Nhâ ̣p vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
n, m = map(int, input().split())

a = []

for i in range(n):
    row = list(map(int, input().split()))
    a.append(row)

for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:4}", end="")
    print()
```

W6A19
Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất
trong list

```python
def vitri(n):
    s=max(n)
    for i in range(len(n)):
        if n[i]==s:
            return i
n =list(map(int,input().split()))
print(vitri(n))
```

W6A20
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1

```python
def vi_tri(n,k):
    for i in range(len(n)):
        if n[i]==k:
            return i
    else:
        return -1
n=list(map(int,input().split()))
k=int(input())
print(vi_tri(n,k))
    
        
```

W6A21
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
s = input().split()
d = {}

for i, w in enumerate(s):
    d[w] = d.get(w, []) + [i]

d = {k: tuple(v) for k, v in d.items()}
print(d)
```

W6A24
Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
def day_so_0(n):
    s=()
    h=()
    for i in n:
        if i==0:
            s+=(i,)
        else:
            h+=(i,)
    return s+h
n=list(map(int,input().split()))
print(day_so_0(n))
        
```

```python
def day_0(a):
    i = 0
    for j in range(len(a)):
        if a[j] != 0:
            a[i], a[j] = a[j], a[i]
            i += 1
a=list(map(int,input().split()))
print(day_0(a))
```

W6A26
Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2
phần tử đầu và cuối

```python
def tong_cuc_dai(a):
    s = 0
    for i in range(1, len(a) - 1):
        if a[i] > a[i - 1] and a[i] > a[i + 1]:
            s = s + a[i]
    return s
```

W6A28
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
thực sự, hay vô danh (không tăng, không giảm)

```python
def ktra_day(a):
    tang = True
    giam = True

    for i in range(1, len(a)):
        if a[i] <= a[i - 1]:
            tang = False
        if a[i] >= a[i - 1]:
            giam = False

    if tang:
        print("Tang thuc su")
    elif giam:
        print("Giam thuc su")
    else:
        print("Vo danh")
```
