```python
import math

a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))

if a + b > c and a + c > b and b + c > a:
    
    if a == b == c:
        loai = "Tam giác đều"
    elif a == b or b == c or a == c:
        loai = "Tam giác cân"
    elif a*a + b*b == c*c or a*a + c*c == b*b or b*b + c*c == a*a:
        loai = "Tam giác vuông"
    else:
        loai = "Tam giác thường"

    p = a + b + c
    s = math.sqrt(p*(p-a)*(p-b)*(p-c))   # công thức Heron

    print(loai)
    print("Chu vi:", p)
    print("Diện tích:", s)
else:
    print("Khong la ba canh cua mot tam giac")
    
```

```python
float(input("Nhập a: "))
float(input("Nhập b: "))
float(input("Nhập c: "))
delta = b**2 - 4*a*c
if delta > 0:
    x1 = (-b + math.sqrt(delta)) / (2*a)
    x2 = (-b - math.sqrt(delta)) / (2*a)
    print("Nghiệm x1:", x1)
    print("Nghiệm x2:", x2)
elif delta == 0:
    x1 = x2 = -b / (2*a)
    print("Nghiệm kép x1 = x2:", x1)
else:
    x1 = x2 = None
    print("Phương trình vô nghiệm")
```

bài 8

```python
a = float(input("nhập a: "))
b = float(input("nhập b: "))
c = float(input("nhập c: "))
d = float(input("nhập d: "))
m = float(input("nhập m: "))
n = float(input("nhập n: "))
det = a*d - b*c
if det != 0:
    x = (d*m - b*n) / det
    y = (a*n - c*m) / det
    print("Nghiệm của hệ phương trình là: x =", x, ", y =", y)
else:
    print("Hệ phương trình vô nghiệm hoặc có vô số nghiệm")
```

```python
x0 = float(input("x0: "))
y0 = float(input("y0: "))
r = float(input("nhập bán kính :"))
x = float(input("nhập tọa độ điểm M(x, y):"))
y = float(input("y: "))
khoang_cach = ((x - x0)**2 + (y - y0)**2)**0.5
if khoang_cach < r:
    print("Điểm M nằm trong đường tròn")
elif khoang_cach == r:
    print("Điểm M nằm trên đường tròn")
else:
    print("Điểm M nằm ngoài đường tròn")
```
