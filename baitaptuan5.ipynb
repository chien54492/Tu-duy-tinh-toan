W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def is_prime(n: int) -> bool:
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

n = int(input())

if is_prime(n):
    print("YES")
else:
    print("NO")
```

W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 

```python
def so_hoan_hao(n: int) -> bool:
    if n < 1:
        return False
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong == n
n = int(input())
if so_hoan_hao(n):
    print("YES")
else:
    print("NO")

```

W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

```python
def giaithua(n):
    if n < 0:
        return None
    s = 1
    for i in range(1, n + 1):
        s *= i
    return s
n = int(input("Nhập n: "))
print(giaithua(n))
```

W5A7
Viết hàm giả lập một máy tính bỏ túi.

```python
def maytinh(num1, num2, operator):
    if operator == '+':
        s = num1 + num2
    elif operator == '-':
        s = num1 - num2
    elif operator == '*':
        s = num1 * num2
    elif operator == '/':
        if num2 == 0:
            return "Không chia được cho 0"
        s = num1 / num2
    else:
        return "Toán tử không hợp lệ"
    
    return round(s, 2)


a = float(input("Nhập số 1: "))
b = float(input("Nhập số 2: "))
op = input("Nhập toán tử (+ - * /): ")

print(maytinh(a, b, op))
```

W5A8
Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng
của chúng.......Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn.

```python
def hamming_distance(x, y):
    return bin(x ^ y).count('1')

x = int(input("Nhập x: "))
y = int(input("Nhập y: "))

print(hamming_distance(x, y))
```

W5A9 Viết hàm trả về tổng các chữ số của số nguyên dương n 

```python
def sum(n):
    t = 0
    while n > 0:
        t += n % 10
        n //= 10
    return t

n = int(input("Nhập n: "))
print(sum(n))
```

W5A10
Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh
xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự
được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có
thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành
'z', 'b' thành 'b' và 'c' thành 'x'.
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai
từ đầu vào có cùng độ dài. 

```python
def dangcau(a, b):
    map_a_b = {}
    map_b_a = {}

    for i in range(len(a)):
        if a[i] in map_a_b:
            if map_a_b[a[i]] != b[i]:
                return False
        else:
            map_a_b[a[i]] = b[i]

        if b[i] in map_b_a:
            if map_b_a[b[i]] != a[i]:
                return False
        else:
            map_b_a[b[i]] = a[i]

    return True

a = input("Nhập chuỗi a: ")
b = input("Nhập chuỗi b: ")

print(dangcau(a, b))
```

W5A12
Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và
0 <= |r| < |divisor|

```python
def chiadu(t, s):
    if s == 0:
        return "Không chia được"
    q = t // s
    r = t % s
    return (q, r)

a = int(input())
b = int(input())

print(chiadu(a, b))
```

W5A13 Viết hàm max_of_two(a, b) trả về số lớn hơn trong hai số

```python
def max_of_two(a, b):
    return a if a > b else b

a = int(input("Nhập a: "))
b = int(input("Nhập b: "))

print(max_of_two(a, b))
```

W5A14 Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op

```python
def compute(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        if b == 0:
            return "Không chia được"
        return a / b
    else:
        return "Sai toán tử"

a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
op = input("Nhập phép toán: ")

print(compute(a, b, op))
```

W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def tamgiacv(a, b, c):
    sides = sorted([a, b, c])
    return sides[0]**2 + sides[1]**2 == sides[2]**2

a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))

print(tamgiacv(a, b, c))
```
