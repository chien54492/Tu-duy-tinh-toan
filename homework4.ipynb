W4A29

```python
a = str(input())
f=""
for i in range(0, len(a)):
    if 'a'<=a[i]<='z':
        f+=chr(ord(a[i])-32)
    elif 'A'<=a[i]<='Z':
        f+=chr(ord(a[i])+32)
    else:
        f+=a[i]
print(f)
```
